```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.title('Box Plot of Heart Dataset')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()
```

```python
df.iloc[:, :5].plot()
plt.title('Line Chart of First 5 Features')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend(loc='best')
plt.show()
```

```python
df.iloc[:, :5].plot.area(alpha=0.5)
plt.title('Area Chart of First 5 Features')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend(loc='best')
plt.show()
```
